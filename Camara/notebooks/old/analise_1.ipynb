```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime
```

```python
def chose_deputado(deputado_escolhido):
    voto_deputado_escolhido = df_votacao[df_votacao['nome'] == deputado_escolhido]
    vde=voto_deputado_escolhido
    vde=vde[vde['voto']!='NaN']
    vde['apoio'] = np.where(vde['voto']==vde['orientacao_gov'],1,0)
    vde['data']= pd.to_datetime(vde['data'])
    vde = vde.sort_values(by='data')
    vde_final = pd.DataFrame()
    vde_final['data']=vde['data']
    vde_final['apoio']=vde['apoio']
    
    return(vde_final)

def chose_partido(partido):
    voto_p = df_votacao[df_votacao['partido'] == partido]
    vp=voto_p
    vp=vp[vp['orientacao_partido']!='NaN']
    vp['apoio'] = np.where(vp['orientacao_partido']==vp['orientacao_gov'],1,0)
    vp['data']= pd.to_datetime(vp['data'])
    vp = vp.sort_values(by='data')
    vp_final = pd.DataFrame()
    vp_final['data']=vp['data']
    vp_final['apoio']=vp['apoio']
    vp_final['partido']=vp['partido']
    
    return(vp_final)
```

```python
df=chose_deputado('Rodrigo Maia')
fig,(axis1) = plt.subplots(1,1,figsize=(15,4))
df.index = df['data']
# df.resample('M').sum()

df=df.drop(columns='data')
df = df.groupby(by=[df.index.year, df.index.quarter]).mean()
df['y/q']=df.index
df.index=df['y/q']

sns.barplot(x=df.index, y=df['apoio'], alpha=.5,ax=axis1).set_title("Support Rate")

```
