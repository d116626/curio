```python
import pandas as pd
partido = 'PSDB'
```

```python
voto_final= pd.read_csv('proposicoes_votos_orientacoes.csv')
voto_final.head(2)
```

```python
voto_final_partido= pd.read_csv('voto_deputados_%s.csv'%(partido))
voto_final_partido.head(2)
```

```python
cols = voto_final.columns.tolist()
print(cols)
```

```python
df = pd.merge(voto_final,voto_final_partido, on=['id_proposicao', 'tipo_proposicao_sigla', 'numero_proposicao', 'ano_proposicao', 'resumo_votacao', 'objeto_votacao', 'cod_sessao', 'data_votacao', 'hora_votacao', 'id_deputado', 'nome', 'uf', 'sigla_partido', 'voto'], how='left')
```

```python
df.head()
```

```python
df.to_csv('proposicoes_votos_orientacoes.csv', encoding='utf-8', index=False)
```
